## Normal Image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def image_negative(image):
    # Calculate image negative by subtracting pixel values from 255
    negative_image = 255 - image
    return negative_image

# Load grayscale image
image = cv2.imread('img.png', 0)  # Use 0 for grayscale
negative_image = image_negative(image)

# Display original and negative image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.subplot(1, 2, 2)
plt.imshow(negative_image, cmap='gray')
plt.title('Negative Image')
plt.show()
```

```python
def plot_histogram(image, title):
    plt.hist(image.ravel(), 256, [0,256])
    plt.title(title)
    plt.show()

# Plot histograms
plot_histogram(image, "Original Image Histogram")
plot_histogram(negative_image, "Negative Image Histogram")
```

## High-Contrast Image

```python
def change_contrast_opencv(image, alpha=1.5, beta=0):
    # Adjust contrast and brightness using addWeighted
    new_image = cv2.addWeighted(image, alpha, np.zeros(image.shape, image.dtype),beta, 0)
    return new_image
```

```python

image = cv2.imread('img.png', 0)
# Apply increased contrast
increased_contrast_image = change_contrast_opencv(image)
negative_image_high_contrast = image_negative(increased_contrast_image)

# Display the result
plt.subplot(1, 2, 1)
plt.imshow(increased_contrast_image, cmap='gray')
plt.title('Increased Contrast Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(negative_image_high_contrast, cmap='gray')
plt.title('Increased Contrast Image (OpenCV)')
plt.axis('off')

plt.show()
```

```python

# Plot histograms
plot_histogram(increased_contrast_image, "High Contrast Image Histogram")
plot_histogram(negative_image_high_contrast, "Negative of High Contrast Image Histogram")
```

## Low Contrast Image

```python
image = cv2.imread('img.png', 0)
decreased_contrast_image = change_contrast_opencv(image=image, alpha=0.5)
negative_image_low_contrast = image_negative(decreased_contrast_image)

# Display the result
plt.subplot(1, 2, 1)
plt.imshow(decreased_contrast_image, cmap='gray')
plt.title('Decreased Contrast Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(negative_image_low_contrast, cmap='gray')
plt.title('Increased Contrast Image (OpenCV)')
plt.axis('off')

plt.show()
```

```python

# Plot histograms
plot_histogram(decreased_contrast_image, "Low Contrast Image Histogram")
plot_histogram(negative_image_low_contrast, "Negative of Low Contrast Image Histogram")
```
