```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def log_transform(image, c):
    """
    Apply log transformation to an image.

    Parameters:
        image (ndarray): Grayscale image.
        c (float): Scaling constant.
    
    Returns:
        ndarray: Log-transformed image.
    """
    # Normalize the image to range [0, 1]
    normalized_image = image / 255.0
    # Apply the log transformation
    log_image = c * np.log(1 + normalized_image)
    # Scale back to range [0, 255]
    log_image = np.uint8(log_image * 255 / log_image.max())
    return log_image

# Load a grayscale image with low gray-level values
image = cv2.imread('cityscape.jpg', cv2.IMREAD_GRAYSCALE)

# Apply log transformation
c = 2.5
log_image = log_transform(image, c)

# Display original and log-transformed images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title(f'Log Transformed Image (c={c})')
plt.imshow(log_image, cmap='gray')

plt.show()
```

```python
# Experiment with different values of 'c'
c_values = [0.5, 1, 2, 5]

plt.figure(figsize=(15, 10))
for i, c in enumerate(c_values):
    transformed_image = log_transform(image, c)
    plt.subplot(2, 2, i + 1)
    plt.title(f'Log Transformed (c={c})')
    plt.imshow(transformed_image, cmap='gray')
plt.tight_layout()
plt.show()
```
