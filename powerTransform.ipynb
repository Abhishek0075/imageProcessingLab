```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def power_law_transform(image, gamma, c=1):
    """
    Apply power-law (gamma) transformation to an image.

    Parameters:
        image (ndarray): Grayscale image.
        gamma (float): Gamma value for transformation.
        c (float): Scaling constant.
    
    Returns:
        ndarray: Transformed image.
    """
    # Normalize the image to range [0, 1]
    normalized_image = image / 255.0
    # Apply the power-law transformation
    transformed_image = c * np.power(normalized_image, gamma)
    # Scale back to range [0, 255]
    transformed_image = np.uint8(transformed_image * 255)
    return transformed_image

# Load a grayscale image
image = cv2.imread('img.png', cv2.IMREAD_GRAYSCALE)

# Apply power-law transformation with gamma = 0.5
gamma = 0.5
power_law_image = power_law_transform(image, gamma)

# Display the original and transformed images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title(f'Power-Law Transformed (γ={gamma})')
plt.imshow(power_law_image, cmap='gray')
plt.show()

```

```python
# Experiment with different gamma values
gamma_values = [0.5, 1, 1.5, 2]

plt.figure(figsize=(15, 10))
for i, gamma in enumerate(gamma_values):
    transformed_image = power_law_transform(image, gamma)
    plt.subplot(2, 2, i + 1)
    plt.title(f'Power-Law Transformed (γ={gamma})')
    plt.imshow(transformed_image, cmap='gray')
plt.tight_layout()
plt.show()
```

```python
medical_image = cv2.imread('xray.png', cv2.IMREAD_GRAYSCALE)

images = [medical_image]
titles = ['Medical']
gamma = 0.8

plt.figure(figsize=(15, 10))
for i, img in enumerate(images):
    transformed_image = power_law_transform(img, gamma)
    plt.subplot(1, 2, i * 2 + 1)
    plt.title(f'Original {titles[i]} Image')
    plt.imshow(img, cmap='gray')

    plt.subplot(1, 2, i * 2 + 2)
    plt.title(f'Transformed {titles[i]} Image (γ={gamma})')
    plt.imshow(transformed_image, cmap='gray')

plt.tight_layout()
plt.show()
```
